#Import and prepare data

```python
# list of indicators that relate to the project the most
chosen_indicators_of_gov_spending = ['General government expenditure by function, general public services, percentage of GDP',
       'General government expenditure by function, economic affairs, percentage of GDP',
       'General government expenditure by function, environment protection, percentage of GDP',
       'General government expenditure by function, health, percentage of GDP',
       'General government expenditure by function, education, percentage of GDP']
```

```python
# filtered original dataset to only keep the data that has an indicator from our list
filtered_funding = funding[funding['Indicator'] .isin(chosen_indicators_of_gov_spending)]
# updated indicator values after filtering
filtered_funding.Indicator.unique()
```

```python
filtered_funding
```

```python
# split the dataset by INDICATOR value, creates a newdataset per indicator (5 in total)
filtered_funding_breakdown = [d for _, d in filtered_funding.groupby(['INDICATOR'])]
```

```python
len(filtered_funding_breakdown)
```

# Helper Functions

```python
# function to plot series
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['TIME']
  ys = series['Value']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])
```

#General Public Services

```python
gen_public_services = filtered_funding_breakdown[0]
gen_public_services
```

```python
gen_public_services_filtered = gen_public_services[['Country', 'TIME', 'Value']]
gen_public_services_filtered
```

```python
countries = sorted(gen_public_services_filtered['Country'].unique())
print(countries)
```

```python
# @title Funding Values Over Time By Country (Selection of Countries)

from matplotlib import pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

specific_countries = ["Belgium", "Iceland", "Greece", "United States", "Italy", "Korea","United Kingdom", "Costa Rica"]

# plot a time series for every country
counter = 0
for country in specific_countries:
  country_gen_public_services = gen_public_services_filtered[gen_public_services_filtered['Country'] == country]
  df_sorted = country_gen_public_services.sort_values('TIME', ascending=True)
  _plot_series(df_sorted, country, counter)
  sns.despine(fig=fig, ax=ax)
  plt.xlabel('TIME')
  _ = plt.ylabel('Value')
  counter += 1

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Funding Values Over Time By Country')
```

```python
# @title Funding Values Over Time By Country (Chosen Country)

from matplotlib import pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 6), layout='constrained')

# plot a time series for a chosen country

country = "United Kingdom" # @param ['Australia', 'Austria', 'Belgium', 'Chile', 'Colombia', 'Costa Rica', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'United Kingdom', 'United States']

country_gen_public_services = gen_public_services_filtered[gen_public_services_filtered['Country'] == country]
df_sorted = country_gen_public_services.sort_values('TIME', ascending=True)
_plot_series(df_sorted, country)
sns.despine(fig=fig, ax=ax)
plt.xlabel('TIME')
_ = plt.ylabel('Value')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Funding Values Over Time By Country')
```
